# <a id='toc1_'></a>[Introduciton](#toc0_)


## <a id='toc1_1_'></a>[Analog vs Digital Video](#toc0_)

- Video systems can generally be divided into two categories: analog and digital.
- Analog video represents visual information using continuous electrical signals, while digital video represents visual information using discrete numerical values.
- Modern multimedia systems, computer vision applications, streaming platforms, and video processing pipelines are primarily based on digital video.
- Digital representations can be efficiently stored, transmitted, compressed, and processed computationally.
- Understanding the difference between analog and digital representation is fundamental to digital video processing.

📝 **Docs**:

- Video: [en.wikipedia.org/wiki/Video](https://en.wikipedia.org/wiki/Video)
- Digital Video: [en.wikipedia.org/wiki/Digital_video](https://en.wikipedia.org/wiki/Digital_video)


### <a id='toc1_1_1_'></a>[Analog Signals](#toc0_)

- An analog signal is a continuous signal whose value changes smoothly over time.
- In analog video systems, image information is represented by continuously varying electrical voltages.
- Mathematically, a continuous-time analog signal can be represented as:
$$
x(t)
$$

- where:
  - $x$ represents the signal amplitude,
  - $t$ represents continuous time.
- Because analog signals are continuous, they can theoretically take infinitely many values within a given range.

**Traditional Analog Video Systems:**

- VHS tapes,
- Analog television broadcasting,
- CRT-based video systems,
- Early surveillance cameras.

📝 **Docs**:

- Analog signal: [en.wikipedia.org/wiki/Analog_signal](https://en.wikipedia.org/wiki/Analog_signal)


### <a id='toc1_1_2_'></a>[Sampling and Quantization](#toc0_)

- Computers cannot directly process continuous analog signals.
- Analog signals must be converted into discrete digital representations through a process called digitization.

$$
\text{Digitization} =
\text{Sampling} + \text{Quantization}
$$

#### <a id='toc1_1_2_1_'></a>[Sampling](#toc0_)

- Sampling converts a continuous signal into a discrete set of samples.
- A sampled signal can be represented as:

$$
x[n] = x(nT_s)
$$

- where:
  - $T_s$ is the sampling interval,
  - $n$ is the sample index.
- In digital video systems, sampling occurs in both spatial and temporal domains.
  - Spatial sampling converts a continuous visual scene into discrete pixels.
  - Temporal sampling converts continuous motion into discrete video frames.
- As a result, digital video is fundamentally composed of discrete frames and discrete pixels.

#### <a id='toc1_1_2_2_'></a>[Quantization](#toc0_)

- After sampling, signal amplitudes must be mapped to a finite set of numerical values.
- For example, an 8-bit grayscale image uses intensity values in the range:

$$
[0, 255]
$$

- This means only 256 discrete intensity levels are available.
- In quantization, some information is inevitably lost. This loss is known as quantization error.

### <a id='toc1_1_3_'></a>[Advantages of Digital Video](#toc0_)

- Efficient Storage
  - Digital video can be stored as computer files and efficiently compressed using modern codecs.
  - This enables practical storage of large multimedia datasets.
- Robustness to Noise
  - Digital systems are significantly more resistant to noise and degradation.
  - Unlike analog systems, digital copies can be reproduced with minimal quality loss.
- Computational Processing
  - Digital representation enables video processing using software and algorithms.
  - Operations listed below are all possible because video data is represented numerically:
    - filtering,
    - enhancement,
    - compression,
    - motion estimation,
    - object tracking,
    - machine learning,
- Efficient Transmission
  - Digital video can be transmitted efficiently through communication networks and streaming platforms.
  - Modern services such as online streaming and video conferencing rely entirely on digital video compression and transmission technologies.
- Scalability
  - Digital video can be easily resized, reformatted, transcoded, and adapted to different devices and bandwidth conditions.


## <a id='toc1_2_'></a>[Digital Video](#toc0_)


### <a id='toc1_2_1_'></a>[Video as a Sequence of Frames](#toc0_)

- A digital video is composed of individual images called frames.
- Mathematically, a video sequence can be represented as:

$$
V = \{F_0, F_1, F_2, \dots, F_n\}
$$

- where:
  - $V$ represents the video,
  - $F_i$ represents the $i$-th frame.
- When frames are displayed sequentially at a sufficiently high speed, the human visual system perceives smooth motion.
- This phenomenon is known as persistence of vision.

📝 **Docs**:

- Digital Video: [en.wikipedia.org/wiki/Digital_video](https://en.wikipedia.org/wiki/Digital_video)


### <a id='toc1_2_2_'></a>[Frame Rate (FPS)](#toc0_)

- Frame rate describes the number of frames displayed or captured per second.
- Frame rate is commonly measured in:

$$
\text{FPS} = \text{Frames Per Second}
$$

- Low frame rates may produce visible motion discontinuities or stuttering, especially during fast motion.
- High frame rates provide smoother temporal representation but increase bandwidth and storage demands.

**Common Frame Rates:**


| Frame Rate | Typical Usage                      |
| ---------- | ---------------------------------- |
| 24 FPS     | Cinema                             |
| 30 FPS     | Television and online video        |
| 60 FPS     | Gaming and high-motion video       |
| 120+ FPS   | Slow-motion and high-speed capture |

- Resolution describes the spatial dimensions of a video frame.
- A frame with width $W$ and height $H$ contains $W \times H$ pixels.
- Higher resolutions contain more spatial detail because more pixels are used to represent the visual scene.

### <a id='toc1_2_4_'></a>[Duration](#toc0_)

- Duration represents the total playback time of a video.
- The duration of a video depends on both the number of frames and the frame rate.
- If a video contains $N$ frames and has a frame rate of $f$ FPS, then the duration is:

$$
T = \frac{N}{f}
$$

- where:
  - $T$ is the duration in seconds,
  - $N$ is the total number of frames,
  - $f$ is the frame rate.
- Together, frame rate, resolution, and duration determine the overall size and complexity of a digital video system.


## <a id='toc1_3_'></a>[Anatomy of a Video Frame](#toc0_)

- Understanding how a frame is represented internally is essential for digital video processing.
- Most video algorithms operate directly on pixel data and numerical arrays.


### <a id='toc1_3_1_'></a>[Pixel](#toc0_)

- A pixel (short for picture element) is the smallest addressable unit of a digital image or video frame.
- Each pixel represents visual information at a specific spatial location.
- A pixel located at coordinates $(x, y)$ stores one or more numerical values describing the intensity or color at that location.
- The total number of pixels in a frame depends on its resolution.
- Because every frame may contain millions of pixels, video processing operations are often computationally intensive.

### <a id='toc1_3_3_'></a>[Frame as a Matrix / Tensor](#toc0_)

- A digital frame can be represented mathematically as a matrix or tensor.
- A grayscale frame is commonly represented as a two-dimensional matrix:

$$
F \in \mathbb{R}^{H \times W}
$$

- where:
  - $H$ is the frame height,
  - $W$ is the frame width.

- Color frames contain multiple channels and are typically represented as three-dimensional tensors:

$$
F \in \mathbb{R}^{H \times W \times C}
$$

- where:
  - (H) is the height,
  - (W) is the width,
  - (C) is the number of channels.


## <a id='toc1_4_'></a>[Spatial vs Temporal Information](#toc0_)

- One of the fundamental differences between images and videos is the presence of temporal information.
- Understanding the distinction between spatial and temporal information is essential in digital video processing.
- Many algorithms operate specifically on one of these domains or combine both simultaneously.


### <a id='toc1_4_1_'></a>[Spatial Information](#toc0_)

- Spatial information describes the visual content within an individual frame.
  - object shapes,
  - edges,
  - textures,
  - colors,
  - brightness,
  - spatial patterns.
- Spatial information is defined over the two-dimensional image plane:

$$
(x, y)
$$

- where:
  - $x$ represents the horizontal coordinate,
  - $y$ represents the vertical coordinate.
- Spatial processing analyzes relationships between neighboring pixels within the same frame.

**Examples of spatial operations:**

- image filtering,
- blurring,
- sharpening,
- edge detection,
- contrast enhancement,
- image segmentation.

**Spatial information is highly important in:**

- computer vision,
- image enhancement,
- object recognition,
- feature extraction.

> Because each frame is fundamentally an image, many image processing techniques can also be applied directly to video frames.


### <a id='toc1_4_2_'></a>[Temporal Information](#toc0_)

- Temporal information describes changes that occur across consecutive frames over time.
- Unlike spatial information, which exists within a single frame, temporal information captures:
  - motion,
  - object displacement,
  - scene dynamics,
  - temporal continuity.
- Temporal information is defined along the time axis:

$$
t
$$

- Neighboring frames in a video are often highly correlated because most parts of the scene remain similar between consecutive frames.
- This temporal correlation is fundamental in many video processing applications, including:
  - motion estimation,
  - optical flow,
  - object tracking,
  - action recognition,
  - video compression.
- Temporal processing operations commonly analyze:
  - frame differences,
  - motion vectors,
  - frame-to-frame correlations,
  - object trajectories.


### <a id='toc1_4_3_'></a>[Spatial-Temporal Nature of Video](#toc0_)

- Digital video simultaneously contains both spatial and temporal information.
- A video signal can therefore be modeled as a three-dimensional function:

$$
V(x, y, t)
$$

- where:
  - $x$ represents the horizontal spatial coordinate,
  - $y$ represents the vertical spatial coordinate,
  - $t$ represents time.
- Many advanced video processing algorithms jointly analyze both domains. Examples include:
  - action recognition,
  - motion tracking,
  - video stabilization,
  - background subtraction,
  - activity analysis,
  - autonomous driving systems.

> Because video inherently contains both spatial and temporal dimensions, digital video processing is fundamentally a multidimensional signal processing problem.


## <a id='toc1_5_'></a>[Video File Fundamentals](#toc0_)

- it is important to distinguish between how video data is **stored**, **compressed**, and **encoded**.
- Terms such as *MP4*, *H.264*, and *bitrate* are often used interchangeably, even though they refer to different concepts.
- A video file typically consists of:
  - a container format,
  - one or more encoded video streams,
  - one or more encoded audio streams,
  - optional metadata such as subtitles, timestamps, and chapter information.


### <a id='toc1_5_1_'></a>[Containers](#toc0_)

- A **container** is a file format that organizes and stores multimedia data.
- A container does not define how video or audio is compressed. Instead, it acts as a wrapper that stores:
  - video streams,
  - audio streams,
  - subtitles,
  - metadata,
  - synchronization information.

**Common video containers:**

| Container | File Extension |
| --------- | -------------- |
| MP4       | `.mp4`         |
| Matroska  | `.mkv`         |
| AVI       | `.avi`         |
| QuickTime | `.mov`         |
| WebM      | `.webm`        |

### <a id='toc1_5_2_'></a>[Codecs](#toc0_)

- A **codec** (coder-decoder or compressor-decompressor) is a method used to encode and decode multimedia data.
- Codecs determine how raw video data is represented and compressed for storage or transmission.
- During encoding:

$$
\text{Raw Video → Encoder → Compressed Bitstream}
$$

- During playback:

$$
\text{Compressed Bitstream → Decoder → Video Frames}
$$

**Common video codecs:**
| Codec        | Typical Usage                                                                |
| ------------ | ---------------------------------------------------------------------------- |
| H.264 (AVC)  | Streaming, broadcasting, video conferencing                                  |
| H.265 (HEVC) | High-efficiency video compression                                            |
| H.266 (VVC)  | Ultra-high-efficiency compression, 4K/8K video, next-generation broadcasting |
| VP9          | Web video                                                                    |
| AV1          | Modern internet streaming                                                    |
| MPEG-2       | DVD and legacy broadcasting                                                  |

### <a id='toc1_5_3_'></a>[Compression](#toc0_)

- Raw digital video contains enormous amounts of data.
- Consider a Full HD video:
  - Resolution: 1920 × 1080
  - Channels: 3 (RGB)
  - Bit depth: 8 bits per channel
  - Frame rate: 30 FPS
- The storage required for uncompressed video can become extremely large, making direct storage and transmission impractical.
- To reduce file size, video data is compressed.
- Compression aims to reduce redundancy while preserving acceptable visual quality.

#### <a id='toc1_5_3_1_'></a>[Lossless Compression](#toc0_)

- Lossless compression preserves all original information.
- Advantages:
  - perfect reconstruction,
  - no quality loss.
- Disadvantages:
  - relatively limited compression ratios.
- Lossless compression is commonly used in archival and professional workflows.

📝 **Docs**:

- Lossless compression: [en.wikipedia.org/wiki/Lossless_compression](https://en.wikipedia.org/wiki/Lossless_compression)


#### <a id='toc1_5_3_2_'></a>[Lossy compression](#toc0_)

- Lossy compression removes information that is considered less perceptually important.
- Advantages:
  - significantly smaller file sizes,
  - efficient streaming and storage.
- Disadvantages:
  - some information is permanently lost.
- Most consumer video formats use lossy compression because it provides a practical balance between quality and file size.

📝 **Docs**:

- Lossy compression: [en.wikipedia.org/wiki/Lossy_compression](https://en.wikipedia.org/wiki/Lossy_compression)


### <a id='toc1_5_4_'></a>[Bitrate](#toc0_)

- Bitrate measures the amount of data used to represent a video stream over time.
- It is typically expressed in:
  - bits per second (bps),
  - kilobits per second (kbps),
  - megabits per second (Mbps).

- Mathematically:

$$
\frac{\text{Total Number of Bits}}{\text{Duration}}
$$

- A higher bitrate generally allows more information to be stored for each frame, often resulting in better visual quality.
- Conversely, lower bitrates reduce storage and bandwidth requirements but may introduce compression artifacts.

**Typical bitrate examples:**
| Application             | Typical Bitrate |
| ----------------------- | --------------- |
| Video conferencing      | 1–5 Mbps        |
| Full HD streaming       | 5–10 Mbps       |
| 4K streaming            | 15–50 Mbps      |
| Professional production | 100+ Mbps       |

## <a id='toc1_6_'></a>[Color Spaces Overview](#toc0_)

- A color space is a mathematical model used to represent colors as numerical values.
- The choice of color space can significantly affect:
  - storage efficiency,
  - compression performance,
  - computational complexity,
  - visual quality.


### <a id='toc1_6_1_'></a>[Grayscale](#toc0_)

- A grayscale image contains only intensity information and does not represent color.
- Each pixel is represented by a single value that indicates brightness.
- Because only one channel is stored, grayscale images require significantly less memory than color images.

**Grayscale representations are frequently used in:**

- edge detection,
- image segmentation,
- feature extraction,
- motion analysis,
- computer vision applications.

### <a id='toc1_6_2_'></a>[RGB](#toc0_)

- The RGB model represents colors using three channels:
  - Red (R),
  - Green (G),
  - Blue (B).
- RGB is an additive color model. Different colors are produced by combining varying amounts of red, green, and blue light.

### <a id='toc1_6_3_'></a>[BGR](#toc0_)

- BGR is identical to RGB except for the channel ordering.
- Instead of:

$$

(R, G, B)

$$


- pixels are stored as:

$$

(B, G, R)

$$

> BGR is particularly important because many computer vision libraries, especially the software library OpenCV, use BGR as their default color format.


### <a id='toc1_6_4_'></a>[YUV](#toc0_)

- The YUV color space separates brightness information from color information.
- It consists of three components:
  - $Y$: luminance (brightness),
  - $U$: blue chrominance component,
  - $V$: red chrominance component.
- The key idea behind YUV is that human vision is generally more sensitive to brightness variations than to color variations.
- By separating luminance and chrominance, video systems can reduce the amount of color information that must be stored or transmitted while maintaining good perceptual quality.

### <a id='toc1_6_5_'></a>[YCbCr](#toc0_)

- YCbCr is a digital color representation derived from RGB and conceptually related to YUV.
- It consists of:
  - $Y$: luma component,
  - $Cb$: blue-difference chroma component,
  - $Cr$: red-difference chroma component.
- Although YCbCr is often informally referred to as YUV, the two are not identical.
- YUV originated in analog television systems, whereas YCbCr was specifically designed for digital image and video representation.
- YCbCr is the dominant color space in modern digital video systems and is widely used in:
  - digital cameras,
  - image compression standards,
  - video compression standards,
  - multimedia storage formats.

## <a id='toc1_8_'></a>[Challenges in Video Processing](#toc0_)


### <a id='toc1_8_1_'></a>[Storage Requirements](#toc0_)

- Video data consumes substantial storage space because it consists of many individual frames.
- The storage required for a video depends on several factors:
  - resolution,
  - frame rate,
  - duration,
  - color depth,
  - compression method.
- As video resolution and frame rate increase, storage demands increase proportionally.
- Modern formats such as 4K and 8K video require significantly more storage capacity than traditional HD formats.


### <a id='toc1_8_2_'></a>[Computational Complexity](#toc0_)

- Video processing algorithms often perform large numbers of operations on every pixel of every frame.
- Many video processing tasks require significantly more computation than simple pixel manipulation, including:
  - filtering,
  - feature extraction,
  - object detection,
  - motion estimation,
  - optical flow,
  - deep learning inference.
- The computational cost increases further with:
  - higher resolutions,
  - higher frame rates,
  - longer video sequences,
  - more sophisticated algorithms.
- As a result, efficient algorithm design and hardware acceleration play important roles in practical video processing systems.


### <a id='toc1_8_3_'></a>[Real-Time Constraints](#toc0_)

- Many applications require processing video as it is being captured or streamed.
- Examples include:
  - video conferencing,
  - autonomous vehicles,
  - surveillance systems,
  - robotics,
  - augmented reality,
  - live broadcasting.
- If processing exceeds this time budget, frames may be delayed or dropped, reducing system responsiveness and visual quality.
- Meeting real-time requirements often requires:
  - optimized algorithms,
  - parallel processing,
  - hardware acceleration,
  - efficient memory management.
- Balancing accuracy and speed is a central challenge in real-time video processing.


### <a id='toc1_8_4_'></a>[Noise and Artifacts](#toc0_)

- Real-world video data is rarely perfect.
- Video signals frequently contain unwanted distortions that degrade visual quality and complicate analysis.
- Noise refers to random variations in pixel values that do not correspond to actual scene content.
- Common sources of noise include:
  - image sensors,
  - low-light conditions,
  - electronic interference,
  - transmission errors.
- Artifacts are distortions introduced by imaging systems or processing operations.
- Examples include:
  - compression artifacts,
  - blocking effects,
  - ringing artifacts,
  - color distortions,
  - motion blur.
- Robust video processing systems must account for these imperfections and often include dedicated enhancement or restoration techniques.


### <a id='toc1_8_5_'></a>[Motion Complexity](#toc0_)

- Motion is one of the defining characteristics of video and one of its greatest challenges.
- Real-world scenes often contain complex motion patterns resulting from:
  - moving objects,
  - camera motion,
  - lighting changes,
  - occlusions,
  - object interactions.
- Motion can occur at different speeds and directions simultaneously within the same scene.
- Examples include:
  - pedestrians crossing a street,
  - vehicles moving in traffic,
  - camera panning during recording,
  - sports events involving rapid movement.
- Accurately modeling motion is difficult because objects may:
  - change appearance,
  - become partially hidden,
  - enter or leave the scene,
  - overlap with one another.
- Many advanced video processing tasks depend heavily on understanding motion, including:
  - object tracking,
  - video stabilization,
  - action recognition,
  - motion estimation,
  - autonomous navigation.
